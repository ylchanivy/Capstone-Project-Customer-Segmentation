##  Explore Sales Pattern at `Item Level`

```python
# Groupby 'StockCode' to agg purchase pattern informaation for stockcode
stockcode=sales_df.groupby('StockCode').agg({
                                            
                                            'Item Description':'first',
                                            'Quantity':'sum',
                                            'Customer ID':pd.Series.nunique,
                                            'Invoice':pd.Series.nunique,
                                            'Country':pd.Series.nunique,
                                            'Revenue':'sum'
                                            })
stockcode.reset_index(inplace= True)
stockcode.head()
```

```python
stockcode.describe()
```

```python
# Top 20 Selling Items by qty
Top_10_Qty_Item=stockcode.sort_values(by='Quantity', ascending= False).head(10)
Top_10_Qty_Item
```

```python
# bar chart 
plt.figure(figsize=(16,10))
ax=sns.barplot(x='Quantity',y='Item Description',data=Top_10_Qty_Item,palette="pastel")
plt.title('Top 20 Popular Products, by Quantity',fontsize=16)
ax.set_xlabel('Quantity',fontsize=14)
ax.set_ylabel('Item',fontsize=14)
ax.tick_params(labelsize=15);
```

```python
# Top 20 Selling Items by Unique Customer
Top_10_customer_Item=stockcode.sort_values(by='Customer ID', ascending= False).head(10)
Top_10_customer_Item
```

```python
# bar chart 
plt.figure(figsize=(16,10))
ax=sns.barplot(x='Customer ID',y='Item Description',data=Top_10_customer_Item,palette="GnBu_d")
plt.title('Top 20 Popular Products, by Customer',fontsize=16)
ax.set_xlabel('Number of Customer',fontsize=14)
ax.set_ylabel('Item',fontsize=14)
ax.tick_params(labelsize=15);
```

##  Explore Sales Pattern at `Transaction Level`

```python
# Groupby 'Invoice' to agg Total Sales informaation for a single invoice
transaction=sales_df.groupby('Invoice').agg({'Quantity':'sum',
                                             'Item Description':pd.Series.nunique,
                                            'Revenue':'sum',
                                            'Price':'sum'})

transaction.reset_index(inplace= True)

transaction.head()
```

```python
transaction.describe()
```

<font color=darkblue> Based on the transaction describe statistic, on average, each transaction sells 284 quantity of items from 21 types of items with a revenue of \\$ 469.98.<br>

```python
transaction.hist();
```

```python
# Transactional level: Revenue

fig, ax = plt.subplots()
fig.set_size_inches(11.7, 5)
sns.distplot(transaction['Revenue'],ax=ax,bins=10,rug=True,kde=False)
ax.set_title('Histogram of Revenue Per Transaction',fontsize=16)
```

```python
transaction.sort_values(by='Revenue', ascending= False).head()
```

<font color=darkblue>Transaction with the highest Revenue is Invoice Number 581483 selling 80,995 quantity of 'PAPER CRAFT , LITTLE BIRDIE' with a total revenue of \\$ 168,469.<br>

```python
# Transactional level: Quantity

fig, ax = plt.subplots()
fig.set_size_inches(11.7, 5)
sns.distplot(transaction['Quantity'],ax=ax,bins=10,rug=True,kde=False)
ax.set_title('Histogram of Quantity Per Transaction',fontsize=16)
```

```python
transaction.sort_values(by='Quantity', ascending= False).head()
```

<font color=darkblue> Transaction with the highest quantity sold is Invoice Number 524174 & Invoice Number 518505 both selling 87167 quantity of items 45 types of items with a total revenue of \$ 11880.84.

```python
# Transactional level: Description

fig, ax = plt.subplots()
fig.set_size_inches(11.7, 5)
sns.distplot(transaction['Item Description'],ax=ax,bins=10,rug=True,kde=False)
ax.set_title('Histogram of No of Item Types Per Transaction',fontsize=16)
```

```python
transaction.sort_values(by='Item Description', ascending= False).head()
```

## Explore Sales Pattern at `Country` Level

```python
# Groupby 'Country' to agg purchase pattern informaation for each country
country=sales_df.groupby('Country').agg({
                                            'Invoice':pd.Series.nunique,
                                            'Customer ID':pd.Series.nunique,
                                            'Quantity':'sum',
                                            'Item Description':pd.Series.nunique,
                                            'Revenue':'sum',
                                            'Price':'sum'})


cols=['Invoice','Customer ID','Quantity','Item Description','Revenue','Price']
```

```python
country.describe()
```

<font color=darkblue> Based on the country describe statistic, on average every country has 144 unique customers and had made 902 transactions and purchased 2564 quantity of items from 718 types of item spending \$ 423,776.<br>

### List of Top 5 Countries with Highest Revenue

```python
# Top 5 Countries with highest revenue

country[cols]=country[cols].apply(pd.to_numeric)
country_top_revenue=country.sort_values(['Revenue'], ascending= False).reset_index()


country_top_revenue.head()
```

```python
country_top_revenue.tail()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(13, 6)
sns.barplot(x='Country',y='Revenue',data=country_top_revenue)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="left")
ax.set_title('Sales Revenue by Country',fontsize=16);
```

<font color=darkblue> Highest sales revenue is generated from United Kingdom amounting to \\$ 14,389,234, followed by EIRE (\\$ 616,570), Netherlands (\\$ 554,038),Germany (\\$ 425,019) and France (\\$ 348,768) while lowest sales revenue is generated from Nigeria amounting to \\$ 140.

## Explore Overall Sales Pattern 

###  Countries of Top 20 Invoice Revenue

```python
transaction.sort_values(by='Revenue',ascending=False,inplace=True)
transaction.head()
```

```python
top20_transaction_spending=transaction.head(20)
top20_transaction_spending_invoice=list(top20_transaction_spending['Invoice'])
top20_transaction_spending_revenue=list(top20_transaction_spending['Revenue'])
```

```python
top20_transaction_spending_country=[]
for invoice in top20_transaction_spending_invoice:
    
    country=sales_df.loc[sales_df['Invoice']==invoice,'Country'].iloc[0]
    
    top20_transaction_spending_country.append(country)
print(set(top20_transaction_spending_country))
```

```python
top20_transaction_spending_country_df=pd.DataFrame(
                                    {'Top 20 Invoice':top20_transaction_spending_invoice,
                                     'Top 20 Invoice Revenue':top20_transaction_spending_revenue,
                                    'Top 20 Invoice Revenue Country':top20_transaction_spending_country})
top20_transaction_spending_country_df.sort_values(by='Top 20 Invoice Revenue',ascending=False)
```

<font color=darkblue> Top 20 highest Invoice Reveune are from 4 countries: United Kingdom, Australia, EIRE and Netherlands.
